```python
df = df.rename(columns={'default.payment.next.month': 'def_pay'})
```

```python
df['SEX'] = df.SEX.astype('category')
df['EDUCATION'] = df.EDUCATION.astype('category')
df['MARRIAGE'] = df.MARRIAGE.astype('category')
df['def_pay'] = df.def_pay.astype('category')
df['PAY_0'] = df.PAY_0.astype('category')
df['PAY_2'] = df.PAY_2.astype('category')
df['PAY_3'] = df.PAY_3.astype('category')
df['PAY_4'] = df.PAY_4.astype('category')
df['PAY_5'] = df.PAY_5.astype('category')
df['PAY_6'] = df.PAY_6.astype('category')
```

## Feature Selection##

```python
df = df.drop(['ID'], axis=1)
x1 = df.drop(['def_pay'], axis=1)
y1 = df['def_pay']
x1.head()
```

## Modelling##

```python
#classifier = DecisionTreeClassifier(max_depth=10, random_state=14) 
```

```python
classifier = DecisionTreeClassifier(random_state = 14, criterion = 'gini', max_depth=10).fit(x1,y1)
```

## Cross Validation##

```python
# Define Scores to Measure
scorer = MultiScorer({
    'ROC AUC'   : (roc_auc_score , {}),
    'Accuracy'  : (accuracy_score , {}),
    'F1 Score'  : (f1_score       , {'average':'weighted'}),
    'Precision' : (precision_score, {'average':'weighted'}),
    'Recall'    : (recall_score   , {'average':'weighted'})
})

start = time.time()
_ = cross_val_score(classifier, x1, y1,scoring=scorer, cv=5)
results = scorer.get_results()

for metric_name in results.keys():
    average_score = np.average(results[metric_name])
    print('%s : %f' % (metric_name, average_score))

print('time', time.time() - start, '\n\n')
```

```python
# Selecting attribute names that will be dimensionally reduced
features = ['PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6', 
            'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6', 
            'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']
others   = ['LIMIT_BAL', 'SEX', 'EDUCATION', 'MARRIAGE', 'AGE', 'def_pay']

# Separating out the features
x2 = df.loc[:, features].values
# Separating out the target
y2 = df.loc[:, others].values

# Executing the PCA
pca = PCA(n_components=6)
principalComponents = pca.fit_transform(x2)
principalDf = pd.DataFrame(data = principalComponents, columns = ['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6'])

# Combine the PCA results with the remaining dataset.
pcadf = pd.concat([df[others], principalDf], axis = 1)
pcadf.head(10)
```

```python
x3 = pcadf.drop(['def_pay'], axis=1)
y3 = pcadf['def_pay']
x3.head()
```

## Modeling##

```python
#classifier = DecisionTreeClassifier(max_depth=10, random_state=14)
#classifier.fit(x3, y3)
```

```python
classifier = DecisionTreeClassifier(random_state = 14, criterion = 'gini', max_depth=10).fit(x3,y3)
```

```python
# Define Scores to Measure
scorer = MultiScorer({
    'ROC AUC'   : (roc_auc_score , {}),
    'Accuracy'  : (accuracy_score , {}),
    'F1 Score'  : (f1_score       , {'average':'weighted'}),
    'Precision' : (precision_score, {'average':'weighted'}),
    'Recall'    : (recall_score   , {'average':'weighted'})
})

start = time.time()
_ = cross_val_score(classifier, x3, y3,scoring=scorer, cv=5)
results = scorer.get_results()

for metric_name in results.keys():
    average_score = np.average(results[metric_name])
    print('%s : %f' % (metric_name, average_score))

print('time', time.time() - start, '\n\n')
```
